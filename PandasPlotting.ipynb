# Multiple Plot using subplot

```python
# 2*2 layout=4 plot
plt.figure(figsize=(8,6))
plt.subplot(221)
plt.plot(d['data1'])

plt.subplot(222)
plt.bar(range(4),d['data2'])

plt.subplot(223)
plt.hist(d['data3'],color='red')

plt.subplot(224)
plt.plot(d['data4'],'--ko')

plt.suptitle('All plot')
```
