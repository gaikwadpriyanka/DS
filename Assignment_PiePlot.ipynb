```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.random((5,4))
```

```python
df = pd.DataFrame(np.random.random((5,4)),columns=['d1','d2','d3','d4'])
df
```

# **Pie Plot**

```python
d2 = pd.Series([10,20,30],name='Central')
d2.plot(kind='pie',title='Data')
```

```python
data = df["d1"]
data
```

```python
d2 = pd.Series([data],name='Central').sum()
d2.plot(kind='pie',autopct='%1.1f%%',title='Data')
```

```python
df.plot.pie(y='d2',figsize=(5,3),autopct='%1.1f%%')
```

```python
df.plot.pie(y='d2',figsize=(5,3),autopct='%1.1f%%')
```

```python
# Transpose
d2 = pd.Series([data],name='Central').sum()
d2.T.plot(kind='pie',autopct='%1.1f%%',title='Data')
```
